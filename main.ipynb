```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_unipolar_nrz(bits):
    signal = np.repeat(bits, 2)
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-0.5, 1.5)
    plt.title('Unipolar NRZ')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
def plot_polar_nrz(bits):
    signal = np.repeat(bits, 2) * 2 - 1
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.title('Polar NRZ')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
def plot_nrz_inverted(bits):
    signal = np.repeat(bits, 2) * 2 - 1
    signal = np.where(signal == 1, -1, 1)
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.title('NRZ Inverted')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
def plot_bipolar_encoding(bits):
    last_level = -1
    signal = []
    for bit in bits:
        if bit == 1:
            last_level = -last_level
            signal.extend([last_level, last_level])
        else:
            signal.extend([0, 0])
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.title('Bipolar Encoding')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
def plot_manchester_encoding(bits):
    signal = []
    for bit in bits:
        if bit == 1:
            signal.extend([1, -1])
        else:
            signal.extend([-1, 1])
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.title('Manchester Encoding')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
def plot_differential_manchester_encoding(bits):
    last_level = 1
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([last_level, -last_level])
        else:
            last_level = -last_level
            signal.extend([last_level, -last_level])
    time = np.arange(len(signal))
    plt.step(time, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.title('Differential Manchester Encoding')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
# Example usage
bits = [1, 0, 1, 1, 0, 0, 1, 0]
plot_unipolar_nrz(bits)
plot_polar_nrz(bits)
plot_nrz_inverted(bits)
plot_bipolar_encoding(bits)
plot_manchester_encoding(bits)
plot_differential_manchester_encoding(bits)
```
